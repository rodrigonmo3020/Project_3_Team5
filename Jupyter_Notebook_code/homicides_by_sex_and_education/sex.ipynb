```python
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../data_sets/sexData.csv',encoding='cp1252')
data
```

```python
data = data.drop(columns=['Unnamed: 5'])
data
```

```python
data = data.dropna()
data
```

```python
columns = list(data.columns)
columns
```

```python
data[columns] = data[columns].replace({',':''}, regex = True)
data[columns] = data[columns].astype('int')
data
```

```python
data.dtypes
```

```python
df1 = data[data['Year'] > 2017]
df1 = df1.reset_index()
df1 = df1.drop(columns=['index','Year','Total'])
df1
```

```python
#Setting x axis
x = list(df1.columns)
x
```

```python
y = []
for column in x:
    y.append(df1[column][0])
y
```

```python
#Plotting 2022
plt.figure(figsize=(5,10))
plot1 = plt.pie(y, labels= x,colors=['blue','pink','green'], autopct='%.2f')
plt.title('2022 Homicides by sex', fontsize = 20, pad='10',loc='center')
plt.show()
```

```python
#Plotting 2021
y2021 = []
for column in x:
    y2021.append(df1[column][3])
#Plotting options
plt.figure(figsize=(5,10))
plot1 = plt.pie(y2021, labels= x,colors=['blue','pink','green'], autopct='%.2f')
plt.title('2021 Homicides by sex', fontsize = 20, pad='10',loc='center')
plt.show()
```
