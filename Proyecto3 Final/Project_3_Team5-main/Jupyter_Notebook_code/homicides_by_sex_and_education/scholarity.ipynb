Importing dependencies 

```python
import pandas as pd 
from matplotlib import pyplot as plt
from pymongo import MongoClient
```

```python
data = pd.read_csv('../data_sets/HomicidiosEscolaridad.csv',encoding='cp1252')
data
```

```python
data = data.fillna(0)
data
```

```python
data.info()
```

```python
data.dtypes
```

```python
columns = list(data.columns)
```

```python
#turning data into the correct dtype
data[columns] = data[columns].replace({',':''}, regex = True)
data[columns] = data[columns].astype('int')
data
```

```python
data.dtypes
```

```python
df1 = data[data['Year'] > 2017]
df1 = df1.reset_index()
df1
```

```python
#Setting x axis
x = columns
x.remove('Unnamed: 15')
x
```

```python
#Setting y axis
y = []
for column in columns:
    y.append(df1[column][0])
y
```

```python
#Plotting 2022
plt.figure(figsize=(15,10))
plt.grid()
plot1 = plt.barh(x,y)
plt.bar_label(plot1)
plt.xlabel('Total of deads')
plt.title('2022 Homicides by study grade', fontsize = 40, pad='30',loc='left')
plt.show()
```
