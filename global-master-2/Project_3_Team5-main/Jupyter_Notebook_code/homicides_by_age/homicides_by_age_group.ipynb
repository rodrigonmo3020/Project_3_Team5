```python
import csv
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt

   
```

```python
with open('./data/Homicidio_edad.csv')as csvfile:
    csvreader= csv.reader(csvfile)

    header= next(csvreader)
```

```python
homicide_age= pd.read_csv('./data/Homicidio_edad.csv')
```

```python
homicide_age
```

```python
homicide_age['total_edad'] = pd.to_numeric(homicide_age['total_edad'].str.replace(',', ''), errors='coerce')
```

```python
data_2018 = homicide_age[homicide_age['año'] == 2018]

plt.figure(figsize=(10, 6))
plt.bar(data_2018['edad'], data_2018['total_edad'], color='blue')

# Adding labels and title
plt.xlabel('Age Group')
plt.ylabel('Total Count')
plt.title('Homicide 2018')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.show()
```

```python
years = sorted(homicide_age['año'].unique())

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(12, 8))

# Iterate through each unique age group and plot a stacked bar for each year
for age_group in homicide_age['edad'].unique():
    data_subset = homicide_age[homicide_age['edad'] == age_group]
    ax.bar(data_subset['año'], data_subset['total_edad'], label=age_group)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.title('Stacked Bar Chart: Homicide Count by Age Group and Year (2018-2022)')

# Adding legend
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the chart
plt.show()
```
